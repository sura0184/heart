```python
import pandas as pd
df = pd.read_csv("/content/heart_disease_data.csv")

numerical_col = 'Year'

mean_val = df[numerical_col].mean()
median_val = df[numerical_col].median()
std_val = df[numerical_col].std()
iqr_val = df[numerical_col].quantile(0.75) - df[numerical_col].quantile(0.25)

print(f"Statistics for '{numerical_col}':")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Standard Deviation: {std_val}")
print(f"IQR (Interquartile Range): {iqr_val}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = mydata['Year']

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

plt.figure(figsize=(10,6))
sns.histplot(data, kde=True, color='green')

plt.axvline(mean, color='blue', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='red', linestyle='-', label=f'Median: {median:.2f}')
plt.axvline(mean + std_dev, color='orange', linestyle=':', label=f'Standard Deviation (+1): {mean + std_dev:.2f}')
plt.axvline(mean - std_dev, color='orange', linestyle=':', label=f'Standard Deviation (-1): {mean - std_dev:.2f}')

plt.title('Distribution of Variable with Mean, Median, and Standard Deviation')
plt.xlabel('Variable')
plt.ylabel('Frequency')


plt.legend()

plt.show()
```
